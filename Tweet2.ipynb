```
counts = tweets.flatMap(lambda line: line.split(" ")) \
             .map(lambda line: line.lower()) \
             .map(lambda word: (word, 1)) \
             .reduceByKey(lambda a, b: a + b)
                      
B = counts.collect()
RT = [item[1] for item in B if item[0] == 'rt'][0]
print "There are", RT, "retweets."
```

```
percentage = float(RT)/float(tweets.count())*100
print "The percentage of retweets to tweets is", percentage, "%."
```
